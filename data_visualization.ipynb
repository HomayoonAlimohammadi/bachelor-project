```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
def warn(*args, **kwargs):
    pass
warnings.warn = warn


htl_etl_params = pd.read_excel('etl_htl_params.xlsx', header=None)
htl_etl_params
```

```python
htl_etl_params_T = htl_etl_params.T
htl_etl_params_T.columns = htl_etl_params_T.iloc[0,:]
htl_etl_params_T = htl_etl_params_T.drop(0, axis=0)
# htl_etl_params_T.reset_index(inplace=True)
htl_etl_params_T
```

```python
perovs_params = pd.read_excel('perovskite_params.xlsx', header=None)
perovs_params
```

```python
perovs_params_T = perovs_params.T
perovs_params_T.columns = perovs_params_T.iloc[0,:]
perovs_params_T = perovs_params_T.drop(0, axis=0)
# htl_etl_params_T.reset_index(inplace=True)
perovs_params_T
```

```python
etl_columns = ['etl_' + '_'.join(column.split()) for column in htl_etl_params_T.columns]
htl_columns = ['htl_' + '_'.join(column.split()) for column in htl_etl_params_T.columns]
perovs_columns = ['perovs_' + '_'.join(column.split()) for column in perovs_params_T.columns]

etl_columns = ['etl_Type'] + etl_columns[1:]
htl_columns = ['htl_Type'] + htl_columns[1:]
perovs_columns = ['perovs_Type'] + perovs_columns[1:]
all_columns = etl_columns + htl_columns + perovs_columns
all_columns
```

```python
import sklearn
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor


X, y = make_regression(n_samples=1000,
                       n_features=30,
                       n_informative=15,
                       n_targets=4,
                       random_state=0,
                       noise=0.5)

print(X.shape, y.shape)
```

```python
lr = LinearRegression()
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor()

lr.fit(X, y)
knn.fit(X, y)
dt.fit(X, y)

print('LinearRegression:', lr.score(X, y))
print('KNN:', knn.score(X, y))
print('DecisionTree:', dt.score(X, y))
```

```python
from numpy import absolute
from numpy import mean
from numpy import std
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold

cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=0)

n_scores = cross_val_score(dt, X, y,
                           scoring='neg_mean_absolute_error',
                           cv=cv, n_jobs=-1)

n_scores = absolute(n_scores)

print(f'MAE: {round(mean(n_scores), 3)} ({round(std(n_scores), 3)})')
```

```python
from sklearn.multioutput import MultiOutputRegressor
from sklearn.svm import LinearSVR


svr = LinearSVR()
wrapper = MultiOutputRegressor(svr)
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=0)

n_scores = cross_val_score(wrapper, X, y,
                           scoring='neg_mean_absolute_error',
                           cv=cv, n_jobs=-1)
n_scores = absolute(n_scores)

print(f'MAE: {round(mean(n_scores), 3)} ({round(std(n_scores), 3)})')
```
